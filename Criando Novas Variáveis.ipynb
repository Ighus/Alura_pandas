## Criando novas váriaveis

```python
import pandas as pd
import matplotlib.pyplot as matplot
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
dados.head(10)
```

```python
# Cria outra coluna com base no calculo de outras duas colunas
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m²'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
# Método de manipular quantas casas decimais irão aparecer
dados['Valor m²'] = dados['Valor m²'].round(2)
```

```python
dados.head(10)
```

```python
dados['Valor Bruto m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa da Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

```python
df_tipo = dados.groupby('Tipo', axis=0).sum()
df_tipo
```

```python
df_tipo['Area'].plot(kind='pie', figsize = (4,4))
matplot.title('Testando uns graficuzinhos')
matplot.show()
```
